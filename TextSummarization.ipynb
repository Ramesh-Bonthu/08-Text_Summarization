## Text Summarization using langchain

```python
## techinque using prompt template

from langchain import PromptTemplate
from langchain import LLMChain

generic_template = """
    Write the summary of a following Speech.
    Speech : {Speech}
    Translate the precise summary to {language}
"""

prompt = PromptTemplate(
    input_variables=['Speech','language'],
    template= generic_template
)

prompt
```

```python
llm_chain = LLMChain(llm=llm,prompt=prompt)
response = llm_chain.run({'Speech':Speech,'language':"Telugu"})
response
```

## Stuff Document Chain Text Summarization

```python
## load the document
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("apjspeech.pdf")

docs = loader.load_and_split()
```

```python
template = """
    Write a concise and short summary of the following speech.
    Speech : {text}
"""

prompt = PromptTemplate(input_variables=['text'],template=template)

```

```python
## summarize library
from langchain.chains.summarize import load_summarize_chain

summary_chain = load_summarize_chain(llm=llm,chain_type="stuff",prompt=prompt,verbose=True)

response = summary_chain.run(docs)

response
```

## Map Reduce Summarization technique

```python
## text Spitters

from langchain_text_splitters import RecursiveCharacterTextSplitter

final_docs = RecursiveCharacterTextSplitter(chunk_size=2000,chunk_overlap=100).split_documents(docs)
```

```python
map_template = """"
    Please summarize the below speech:
    Speech:{text}
    Summary:
"""
map_prompt_chain = PromptTemplate(input_variables=['text'],template=map_template)
```

```python
final_template = """
    Provide the final summary of the speech with these important points.
    Add a motivational titie, start with a precise summary with an introduction and provide the summary in number points for the speech.
    Speech : {text}
"""
final_prompt_chain = PromptTemplate(input_variables = ['text'],template=final_template)
```

```python
summary_chain = load_summarize_chain(
    llm=llm,
    chain_type="map_reduce",
    map_prompt=map_prompt_chain,
    combine_prompt = final_prompt_chain,
    verbose=True   
)

output = summary_chain.run(final_docs)
output
```

## Refine summarization technique

```python
chain = load_summarize_chain(
    llm=llm,
    chain_type='refine',
    verbose=True
)
output = chain.run(final_docs)
output
```
